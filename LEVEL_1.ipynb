```python
data['Aggregate rating'].hist(bins=20)
plt.xlabel('Aggregate Rating')
plt.ylabel('Frequency')
plt.title('Distribution of Aggregate Rating')
plt.show()
```

```python
cuisines = data['Cuisines'].value_counts().head(20)
print("Top cuisines with the highest number of restaurants:")
print(cuisines)
```

```python
cities = data['City'].value_counts().head(20)
print("Top cities with the highest number of restaurants:")
print(cities)
```

```python
Country_Code = data['Country Code'].value_counts()
print("Distribution of Country Code:")
print(Country_Code)
```

```python
restaurant = data['Restaurant Name'].value_counts().head(20)
print("Restaurant Name Distribution:")
print(restaurant)
```

```python
locality = data['Locality'].value_counts().head(20)
print("Locality Distribution:")
print(locality)
```

```python
rating_text = data['Rating text'].value_counts().head(20)
print("Rating Text Distribution:")
print(rating_text)
```

```python
Table_booking = data['Has Table booking'].value_counts().head(20)
print("Table Booking Distribution:")
print(Table_booking)
```

```python
city = data['City'].value_counts()
plt.figure(figsize=(10, 5))
city.head(10).plot(kind='bar', color='purple')
plt.title('Top 15 Cities with the Most Restaurants')
plt.xlabel('City')
plt.ylabel('Number of Restaurants')
plt.xticks(rotation=40)
plt.show()
```

```python
country = data['Country Code'].value_counts()
plt.figure(figsize=(10, 5))
country.head(10).plot(kind='bar', color='blue')
plt.title('Restaurants Distribution Across Countries')
plt.xlabel('Country Code')
plt.ylabel('Number of Restaurants')
plt.xticks(rotation=40)
plt.show()
```

```python
data['Aggregate rating'].value_counts().plot(kind='bar', color ='red')
plt.title('Class Distribution')
plt.xlabel('Rating')
plt.xticks(rotation=60)
plt.ylabel('Count')
```

```python
top_cuisine = data['Cuisines'].value_counts().head()
top_cuisine.plot(kind = 'bar', color ='salmon')
plt.title('Top Cuisines Distribution')
plt.xticks(rotation=45)
plt.show()
```

```python
pune= data[data['City'] == 'Pune']
plt.figure(figsize=(12, 6))
sns.countplot(data=pune,x='Locality' , palette='coolwarm' )
plt.title('Distribution of Restaurants in Pune')
plt.xlabel('Locality')
plt.ylabel('Count')
plt.xticks(rotation=45, ha='right')
plt.show()
```

```python
Correlation = data[['Longitude' , "Latitude" , "Aggregate rating"]].corr()
Correlation
```

```python
plt.figure(figsize=(10, 5))
sns.heatmap(Correlation, annot=True, cmap='viridis', linewidths=5, fmt=".2f")
plt.title('Correlation Between Resturant Location and Rating')
plt.show()
```
