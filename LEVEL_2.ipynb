```python
dataset = dataset.dropna(axis=0)
```

```python
dataset.isnull().sum()
```

```python
table= (dataset['Has Table booking'].value_counts(normalize=True) * 100)
online = (dataset['Has Online delivery'].value_counts(normalize=True) * 100)
print("Percentage of restaurants offering table booking:")
print(table)
print("\nPercentage of restaurants offering online delivery:")
print(online)
```

```python
table = dataset.groupby('Has Table booking')['Aggregate rating'].mean()
print("Average ratings by Table Booking:")
print(table)
```

```python
online_delivery = dataset.groupby('Price range')['Has Online delivery'].value_counts(normalize=True) * 100
print("Percentage of restaurants offering online delivery by price range:")
print(online_delivery)
```

```python
online_delivery.plot(kind='bar')
plt.xlabel('Price Range')
plt.ylabel('Number of Restaurants with Online Delivery')
plt.title('Availability of Online Delivery by Price Range')
plt.show()
```

```python
average_rating = dataset.groupby('Price range')['Aggregate rating'].mean()
print("Average rating for each price range:")
print(average_rating)
```

```python
average_rating = dataset.groupby('Price range')['Aggregate rating'].mean()
plt.figure(figsize=(12, 6))
average_rating.plot(kind='bar', color='maroon')
plt.title('Average Rating for Each Price Range')
plt.xlabel('Price Range')
plt.ylabel('Average Rating')
plt.xticks(rotation=0)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
dataset = pd.get_dummies(dataset, columns=['Has Table booking', 'Has Online delivery'])
print(dataset.head())
```

```python
plt.figure(figsize=(10, 5))
plt.hist(dataset['Aggregate rating'], bins=30, color='yellow', edgecolor='black')
plt.title('Distribution of Ratings')
plt.xlabel('Aggregate Rating')
plt.ylabel('Frequency')
plt.show()

plt.figure(figsize=(10, 5))
sns.countplot(x='Aggregate rating', data=dataset , palette='coolwarm')
plt.title('Count of Each Rating')
plt.xlabel('Aggregate Rating')
plt.ylabel('Count')
plt.show()
```
